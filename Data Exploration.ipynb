In the process of Data Cleaning , 

Lets check whether in dataset the thickness of sub-base course is smaller than base course

and the thickness of base course is smaller than surface course in any of the data

Generally T(sub-base)>T(base)>T(Surface)

```python
condition=df['Base_thickness(mm)']>df['Sub_base_thickness(mm)']
condition
```

```python
count_true=condition.sum()
print('The number of data where the thickness of base course is larger than the thickness of sub-base is:',count_true)
```

```python
#lets update the values where the condition is true
df.loc[condition,'Base_thickness(mm)']=df.loc[condition,'Sub_base_thickness(mm)']
```

```python
df.head()
```

It seems to work :)

Now Lets do same for base and wearing course

```python
condition2=df['Base_thickness(mm)']<df['Surface_Thickness(mm)']
condition2
```

```python
#lets count number of data where surface thickness is greater than base thickness
count2=condition2.sum()
print('The number of data where base thicknes is smaller than surface thickness is',count2)
```

```python
#lets print those data
df.loc[condition2]
```

```python
#lets update the data and make base thickness equal to the surface thickness in those 17 datas
df.loc[condition2,'Base_thickness(mm)']=df.loc[condition2,'Surface_Thickness(mm)']
```

```python
#now lets check the condition again
df.loc[condition2]
```

We can see in the above output the value of base course thickness has been replaced by value of surface thickness where the condition applied.

Now, why not change construction year which is in AD to construction year in BS.

```python
df['Construction_year(AD)']=df['Construction_year(AD)']+57 #Generally there is 57 year differences between AD and BS
```

```python
df['Construction_year(AD)'].head()
```

The value within the column has been updated, however the name of column head still shows AD. klets replace that.

```python
df.rename(columns={'Construction_year(AD)':'Construction_year(BS)'})
df.head()
```

```python
#we can see, it has not been replaced, so what we can do is use inplace=True to replace in original dataframe 
df.rename(columns={'Construction_year(AD)':'Construction_year(BS)'},inplace=True)
df.head()
```
